```python
import nltk
import string
import re, unicodedata
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk import pos_tag
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline
import torch

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)
```

```python
model_name = "mistralai/Mistral-7B-Instruct-v0.1"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
        model_name,
        load_in_4bit=True,
        quantization_config=bnb_config,
        torch_dtype=torch.bfloat16,
        device_map="auto",
        trust_remote_code=True,
    )
```

```python
input_file='<---input file path here--->'
#print(string.punctuation)
file_content=open(input_file,encoding='utf-8').read()
file_content=file_content.lower()
parlist=file_content.split('\n')
#content = "".join([char for char in file_content if ((char not in string.punctuation) or (char=='\n'))])
#print(content)

words = nltk.word_tokenize(file_content)
def strip_ascii(text):
  ans =''
  for word in text:
      new_word = unicodedata.normalize('NFKD', word).encode('ascii','ignore').decode('utf-8','ignore')
      ans+=new_word
  return ans



new_words=[]
for word in words:
  new_word = unicodedata.normalize('NFKD', word).encode('ascii','ignore').decode('utf-8','ignore')
  new_words.append(new_word)
#st=" ".join(word for word in new_words)
tx="\n".join([strip_ascii(ele) for ele in parlist])
print(tx)
```

```python
def preprocess(filename):
    f = open(filename,'r')
    text = f.read()
    text = text.lower()

    text_p = "".join([char for char in text if char not in string.punctuation])

    words = word_tokenize(text_p)

    stop_words = stopwords.words('english')
    filtered_words = [word for word in words if word not in stop_words]

    porter = PorterStemmer()
    stemmed = [porter.stem(word) for word in filtered_words]

    pos = pos_tag(filtered_words)

    return text_p, words, filtered_words
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer = tokenizer,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
textprompt="You are an expert interviewer who interviews a Twitter post of a person. You formulate questions based on quotes from the Twitter posts. Below is one such post. Formulate a question that the post would be the perfect answer to. The question should be short and directed at the author of the post like in an interview. The question is short. Remember, make the question as short as possible. Do not give away the answer in your question and the response should be differently worded than the question. Also: If possible, ask for motivations, feelings, and perceptions rather than events or facts.The generated text should only have the question.Here is some context that might help you formulate the question regarding the quote: ' "
prompt = "You are an expert interviewer who interviews a Twitter post of a person. You formulate questions based on quotes from the Twitter posts. Below is one such post. Formulate a question that the post would be the perfect answer to. The question should be short and directed at the author of the post like in an interview. The question is short. Remember, make the question as short as possible. Do not give away the answer in your question and the response should be differently worded than the question. Also: If possible, ask for motivations, feelings, and perceptions rather than events or facts.The generated text should only have the question.Here is some context that might help you formulate the question regarding the quote: ' " + paragraphs[7]+" '"

sequences = pipe(
    prompt,
    do_sample=True,
    max_new_tokens=100,
    temperature=0.7,
    top_k=50,
    top_p=0.95,
    num_return_sequences=1,
)
```
